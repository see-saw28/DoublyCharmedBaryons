# Omegaccp

```python
obs_min = 0
obs_max = 500
obs_bin_width = 4
obs_bin = int((obs_max-obs_min)/obs_bin_width)
```

```python
def plot_data(data):
    fig, ax = plt.subplots()
    # histogram with centered bins
    histo = hist.Hist(hist.axis.Regular(obs_bin, obs_min, obs_max, label='Observable'))
    histo.fill(data)
    # errorbar histogram for the data
    ax.errorbar(histo.axes.centers[0], histo.values(), xerr=histo.axes.widths[0]/2,
            yerr=np.sqrt(histo.values()), fmt='.', label='Data', color='black')
    # labels
    #ax.set_xlabel('$m_{cand}(\Xi_{cc}^{++})[MeV/c^2]$')
    ax.set_xlabel("$m(\Xi_{cc}^{++} K^-)-m(\Xi_{cc}^{++} )-m(K^-)[MeV/c^2]$")
    ax.set_ylabel(f'Events/( {obs_bin_width} MeV/$c^2$ )')
```

```python
def plot_mass(data, label = None, density = False, bins = obs_bin):
    h, bins = np.histogram(data, bins=bins, range=[obs_min, obs_max])
    mplhep.histplot(h, bins, density=density, yerr=False, label = label)
    plt.xlabel("$m(\Xi_{cc}^{++} K^\pm)-m(\Xi_{cc}^{++} )-m(K^\pm)[MeV/c^2]$")
    plt.ylabel(f'Events/( {obs_bin_width} MeV/$c^2$ )')
    plt.xlim(bins[0], bins[-1])
```

## Loose selection

```python
Omegaccp_RS_Loose_data_cut2 = Omegaccp_RS_Loose_data_cut.query("MLP>0.22")
Omegaccp_WS_Loose_data_cut2 = Omegaccp_WS_Loose_data_cut.query("MLP>0.22")
Omegaccp_SB_Loose_data_cut2 = Omegaccp_SB_Loose_data_cut.query("MLP>0.22")

# Omegaccp_RS_Loose_data_cut2 = Omegaccp_RS_Loose_data_cut.query("BDT>-0.05")
# Omegaccp_WS_Loose_data_cut2 = Omegaccp_WS_Loose_data_cut.query("BDT>-0.05")
```

```python
plot_data(Omegaccp_RS_Loose_data_cut2.C_KaonDTF_C_M-4115.1)
```

```python
plot_mass(Omegaccp_RS_Loose_data_cut2.C_KaonDTF_C_M-4115.1 , density = False , label = r"RS")
plot_mass(Omegaccp_WS_Loose_data_cut2.C_KaonDTF_C_M-4115.1 , density = False  , label = r"WS")
plot_mass(Omegaccp_SB_Loose_data_cut2.C_KaonDTF_C_M-4115.1 , density = False  , label = r"SB")
plt.legend(loc='best')
```
